```python
## Checking if there are any empty cells in the columns
Online.isnull().sum()
```

```python
## Removing the Rows containg null values. 1 is columns and 0 is rows. Using inplace = true to update the dataframe 
Online.dropna(how = 'all', axis=0, inplace = True)
Online.dropna(how = 'all', axis=1, inplace = True)
```

```python
## Checking if there are still null values in the columns
Online.isnull().sum()
```

```python
Online.head()
```

```python
## There are extra trailing and leading spaces in column names
Online.columns
```

```python
## Stripping spaces from column names and reassigning the clean column names to the dataset
Online.columns = Online.columns.str.strip()
```

```python
Online.columns
```

```python
## Replacing any spaces with Null values so that we can use fillna method
Online.replace('',None, inplace = True)
```

```python
Online.head(50)
```

```python
## Interchanging the Column names because they were misplaced 
Online.rename(columns = {'Sales':'Product Name','Product Name':'Sales'}, inplace = True)
```

```python
Online.head(50)
```

```python
## Deleting all the empty rows for these columns as they are the important columns for Analysis 
Online = Online.dropna(subset = ['Ship Mode','State','Region','Category','Sub-Category','Sales','Quantity','Profit'])
```

```python
## Checking if the important columns has anymore null values
Online.isnull().sum()
```

```python
## Forward filling the dataset
Online = Online.fillna(method = 'ffill')
```

```python
## No more null values found
Online.isnull().sum()
```

```python
## Removing the column Row ID because it's not needed to us
## Removing the Column Customer Balance because all of the values were Zero 
Online.drop(['Row ID','Customer Balance'], axis=1, inplace = True)
```

```python
Online.head()
```

```python
## Checking if There are any Duplicate Rows 
Duplicates= Online[Online.duplicated()]
```

```python
## There are duplicate rows
Duplicates
```

```python
## Keeping the First Duplicate and Removing the rest of the Duplicates
Online = Online.drop_duplicates( keep='first')
```

```python
## checking to see if there are any more duplicates
Duplicates= Online[Online.duplicated()]
Duplicates
```

```python
## Changing the Data type of Quantity to integer as it is in float
Online['Quantity'] = Online['Quantity'].astype(int)
```

```python
Online.dtypes
```

```python
## changing the Date format of order date and ship date to DD/MM/YYYY
Online['Order Date'] = Online['Order Date'].dt.strftime('%d/%m/%Y')
Online['Ship Date'] = Online['Ship Date'].dt.strftime('%d/%m/%Y')
```
